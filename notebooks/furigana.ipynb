```python
from furigana.furigana import print_html


text = "澱んだ街角で僕らは出会った"
# print_html('澱んだ街角で僕らは出会った')

def generate_transcription(text):
    """
    Generates furigana transcription for a given Japanese text.

    :param text: A string containing Japanese text.
    :return: A string with the original text and its furigana transcription.
    """
    return print_html(text)
```
